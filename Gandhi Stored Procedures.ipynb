```sql
/****** Object:  StoredProcedure [dbo].[BuyItemByCredit]    Script Date: 8/7/2022 5:17:33 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE procedure [dbo].[BuyItemByCredit] @buyerId int, @sellerId int, @itemId int ,@price int 
as
declare @buyerCoins int= (select coinsCredit from Users_proj
							where userId=@buyerId)
declare @sellerCoins int= (select coinsCredit from Users_proj
							where userId=@sellerId)
declare @difference int
begin
--insert to users items for the buyer
	insert into UserItems_proj (userId, ItemId, userSideCode, updateDate) 
	values (@buyerId, @itemId,'B',GETDATE())
--update items status
UPDATE Items_proj
SET itemStatus = 'N'
WHERE ItemId=@itemId
--update buyer coins credit
if(@buyerCoins>=@price)
	begin
	UPDATE Users_proj
	SET coinsCredit = @buyerCoins-@price
	WHERE userId=@buyerId
	set @difference=0; --left to pay
	end
else
	begin
	UPDATE Users_proj
	SET coinsCredit = 0
	WHERE userId=@buyerId
	set @difference=@price-@buyerCoins; --left to pay
	end
--update seller coins credit
	begin
	UPDATE Users_proj
	SET coinsCredit = @sellerCoins+@difference+@difference*0.1
	WHERE userId=@sellerId
	end
return @difference; 
end
GO
```

# [dbo].[InsertTag]

```sql
/****** Object:  StoredProcedure [dbo].[InsertTag]    Script Date: 8/7/2022 5:17:33 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



CREATE procedure [dbo].[InsertTag] @TagDesc nvarchar(20), @itemId int
	as 
	declare @tagId  int, @count int;

	if not exists (select * from Tags_proj where tagDesc=@TagDesc) --tag doesnt exist
	begin --insert new tag
		insert into Tags_proj (tagDesc)
		values (@TagDesc)
		set @tagId= SCOPE_IDENTITY(); --new tag ID
		set @count=1;
	end
	else --tag already exist
	begin
		set @tagId= (select tagId from Tags_proj --existing tag ID
								where tagDesc=@TagDesc)
					set @count=1;
	end
	begin
		insert into ItemTags_proj (tagId, itemId)
		values (@tagId, @itemId)
		set @count=2; 
	end
	return @count
		--insert to connection table 
GO
```

# [dbo].[InsertTags]

```sql
/****** Object:  StoredProcedure [dbo].[InsertTags]    Script Date: 8/7/2022 5:17:33 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[InsertTags] @tags AS dbo.tagsListF READONLY, @ItemId int 
as
begin
	DECLARE
		@Init int= 1,
		@NumRows int
	SELECT @NumRows= COUNT(*) FROM @tags
	WHILE @Init<= @NumRows
	BEGIN
		declare @tmpDesc nvarchar(50)
		declare @tagId int
		set @tmpDesc= (select TagDesc from @tags 						where ID=@Init)
		if not exists (select * from Tags_proj where tagDesc=@tmpDesc)  --tag doesnt exist			begin --insert new tag			insert into Tags_proj (tagDesc)
			values (@tmpDesc)			set @tagId= SCOPE_IDENTITY(); --new tag ID			end		else --tag already exist			begin			set @tagId= (select tagId from Tags_proj --existing tag ID						where tagDesc=@tmpDesc)			end		insert into ItemTags_proj  (tagId, itemId) --insert to connection table 
		values
		(@tagId,@ItemId)
		SET @Init= @Init + 1
	END

end
GO
```

# [dbo].[InsertUsersWatchedItems]

```sql
/****** Object:  StoredProcedure [dbo].[InsertUsersWatchedItems]    Script Date: 8/7/2022 5:17:33 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE procedure [dbo].[InsertUsersWatchedItems] @itemId int, @userId int
as
declare @views int
set @views=(select viewCounter from Items_proj
				where ItemId=@itemId) --current views
begin
if exists (select * from UserWatchItems_proj
			where @itemId=itemId and @userId=userId )
	begin
	UPDATE UserWatchItems_proj SET watchDate = GETDATE() WHERE @itemId=itemId and @userId=userId;
	UPDATE Items_proj SET viewCounter = @views+1 WHERE @itemId=itemId ;
	return 1 
	end
else
	begin 
	insert into UserWatchItems_proj (itemId,userId,watchDate)
	values (@itemId,@userId,getdate())
	UPDATE Items_proj SET viewCounter = @views+1 WHERE @itemId=itemId ;
	return 1
	end
end
GO
```

# [dbo].[MakeTransaction]

```sql
/****** Object:  StoredProcedure [dbo].[MakeTransaction]    Script Date: 8/7/2022 5:17:33 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO










CREATE procedure [dbo].[MakeTransaction] @itemId int, @buyerId int , @sellerId int, @side char(1)
as
declare @transactionNumber int;
declare @updated int=0;
declare @result int;


if not exists (select * from Transactions_proj
				where itemId=@itemId )			
				begin

				if (@side='B')
				begin
				insert into Transactions_proj (itemId, buyerId, buyerConfirm)
				values (@itemId, @buyerId, 'Y')
				set @result=1;
				return @result;
				end
				

				else 
				begin
				insert into Transactions_proj (itemId, sellerId, sellerConfirm, buyerId)
				values (@itemId, @sellerId, 'Y' , @buyerId)
				set @result=1;
				return @result;
				end

				end 
				

else if exists (select * from Transactions_proj where itemId=@itemId and (sellerConfirm is null or buyerConfirm is null))
			begin	
				if (@side='B')  
				begin
				if exists (select transactionId from Transactions_proj
												where itemId=@itemId and (buyerConfirm IS NULL OR buyerConfirm = '') and buyerId=@buyerId)
					begin
					set @transactionNumber= (select transactionId from Transactions_proj
												where itemId=@itemId and (buyerConfirm IS NULL OR buyerConfirm = '') and buyerId=@buyerId)
					update Transactions_proj
					set buyerConfirm='Y'
					where itemId=@itemId and (buyerConfirm IS NULL OR buyerConfirm = '') and buyerId=@buyerId 
					set @updated=1; 
					set @result=2;

					end
				end

				else
				begin
				if exists (select transactionId from Transactions_proj
												where itemId=@itemId and (sellerId IS NULL OR sellerId = '') and buyerId=@buyerId)
					begin
					set @transactionNumber= (select transactionId from Transactions_proj
												where itemId=@itemId and (sellerId IS NULL OR sellerId = '') and buyerId=@buyerId)
					update Transactions_proj
					set sellerId=@sellerId, sellerConfirm='Y'
						where itemId=@itemId and (sellerId IS NULL OR sellerId = '') and buyerId=@buyerId
						set @updated=1;
						set @result=2;

					end
				end
				
				if (@updated=1)
				begin

					insert into UserItems_proj (ItemId, userId, userSideCode, updateDate) 
					values (@itemId, @buyerId, 'B', GETDATE())
					UPDATE Items_proj
					SET itemStatus = 'N'
					WHERE ItemId=@itemId;
					
					declare @seller int= (select sellerId from Transactions_proj
											where transactionId=@transactionNumber)

					update UserItems_proj
					set userSideCode='H'
					where userId=@seller and ItemId=@itemId
					return @result;

				end
		end	

else if exists (select * from Transactions_proj where itemId=@itemId and sellerConfirm='Y' and buyerConfirm='Y')
			
			
				if (@side='B')
				begin
				insert into Transactions_proj (itemId, buyerId, buyerConfirm)
				values (@itemId, @buyerId, 'Y')
				set @result=1;
				return @result;
				end
				

				else 
				begin
				insert into Transactions_proj (itemId, sellerId, sellerConfirm, buyerId)
				values (@itemId, @sellerId, 'Y' , @buyerId)
				set @result=1;
				return @result;
				end

				

GO
```

# [dbo].[SellExistingItem]

```sql
/****** Object:  StoredProcedure [dbo].[SellExistingItem]    Script Date: 8/7/2022 5:17:33 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE procedure [dbo].[SellExistingItem] @itemId int, @userId int
as
begin
if exists(select * from UserItems_proj
			inner join Items_proj
			on UserItems_proj.ItemId=Items_proj.ItemId
			where UserItems_proj.ItemId=@itemId and userId=@userId 
			and userSideCode='B' and Items_proj.itemStatus='N')
		begin
		UPDATE UserItems_proj
		SET userSideCode='S'
		WHERE UserItems_proj.ItemId=@itemId and userId=@userId 
				and userSideCode='B'

		UPDATE Items_proj
		SET itemStatus='A'
		WHERE ItemId=@itemId 

		delete from userInterestedItems_proj
		where userId=@userId and itemId=@itemId

		if exists (select * from UserLikedItems_proj
					where userId=@userId and itemId=@itemId)
			begin
			delete from UserLikedItems_proj
			where userId=@userId and itemId=@itemId
			end
		else
			return 1;

		return 1;
		end
else
		return 0;
end
GO
```
